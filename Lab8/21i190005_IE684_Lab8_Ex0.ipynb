```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
list_z=np.linspace(-10,10)
```

```python
G_h=[max(0,1-z) for z in list_z]
G_l=[math.log(1 + math.exp(-z)) for z in list_z]
G_sh=[(max(0,1-z))**2 for z in list_z]
```

```python
plt.figure(figsize=(9,7))
plt.plot(list_z,G_h)
plt.plot(list_z,G_l)
plt.plot(list_z,G_sh)
plt.legend(["G_h","G_l","G_sh"])
plt.title("Graphs of three different Loss function")
plt.grid()
plt.xlabel("value of z")
plt.ylabel("values of loss functions");
```

In above graph, we can see that the value of all loss functions is zero for positive values of z, and for negative values of z the squared hinge loss has more value as compared to other loss functions.
